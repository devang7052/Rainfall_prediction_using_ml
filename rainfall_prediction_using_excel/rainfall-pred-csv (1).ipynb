```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error


x_tom=df_processed.loc[:,'MinTemp':'RainToday']
y_tom=df_processed.loc[:,'RainTomorrow']
# Normalize features
scaler = MinMaxScaler(feature_range=(0, 1))
X_scaled = scaler.fit_transform(x_tom)

# Reshape input to be [samples, time steps, features]
X_reshaped = np.reshape(X_scaled, (X_scaled.shape[0], 1, X_scaled.shape[1]))

# Splitting into train and test sets
X_train_tom, X_test_tom, y_train_tom, y_test_tom = train_test_split(X_reshaped, y_tom, test_size=0.2, shuffle=False)

# Define LSTM model
model = Sequential()
model.add(LSTM(50, input_shape=(X_train_tom.shape[1], X_train_tom.shape[2])))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train model
model.fit(X_train_tom, y_train_tom, epochs=100, batch_size=1, verbose=1)

# Evaluate model

from sklearn.metrics import mean_squared_error,accuracy_score,recall_score,f1_score
y_pred = model.predict(X_test_tom)
y_pred = np.where(y_pred >= 0.5, 1, 0)
# Calculate accuracy
accuracy = accuracy_score(y_test_tom, y_pred)
f1= f1_score(y_test_tom, y_pred)
recall= recall_score(y_test_tom, y_pred)
print("recall_tomay_lstm:", recall)
print("f1_tomay_lstm:", f1)
print("Accuracy_tomay_lstm:", accuracy)
# recall_tom_lstm: 0.6125356125356125
# f1_tom_lstm: 0.7142857142857143
# Accuracy_tom_lstm: 0.868601986249045
```
